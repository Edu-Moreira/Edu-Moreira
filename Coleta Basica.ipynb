```python
import pandas as pd
import numpy as np
import yfinance as yf
import requests
import zipfile
import io
import xlsxwriter
import time
```

```python
ano = 2023

url = str('https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/itr_cia_aberta_'+str(ano)+'.zip')

response = requests.get(url)

```

```python
# Criando o objeto ZipFile corretamente
zf = zipfile.ZipFile(io.BytesIO(response.content))

# Obtendo a lista de nomes dos arquivos no objeto ZipFile
filenames = zf.namelist()

# Imprimindo os nomes dos arquivos com seus índices
for index, filename in enumerate(filenames):
    print(f"{index}: {filename}")



# Número índice do arquivo desejado
i = 14 

```

```python
arquivo = zf.namelist()[i]

zf = zf.open(arquivo)

lines = zf.readlines()
lines =[i.strip().decode('ISO-8859-1') for i in lines]
lines = [i.split(';') for i in lines]

df = pd.DataFrame(lines[1:], columns = lines[0])
df
```
